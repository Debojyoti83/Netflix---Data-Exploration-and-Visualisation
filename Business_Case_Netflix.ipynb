#**Handling Missing Values**
*   Missing values can hinder accurate analysis. by using .isna().sum() function lets explore the missing values in the dataset.

```python
df.isna().sum()
```

```python
df.isna().sum().sum()
```

* There are a total of 4,307 missing values accross the entire dataset with   
2,634 null values under "director", 825 under "cast", 831 under "country", 10 under "date_added", 4 under "rating", and 3 under "duration".

* Imputation and dropping is the handling method for missing values. Can use mean, mode, or use predictive modeling. Lets handel missing values by using different approach for different column.
* For the "director" and "cast" columns replace missing values with "No information avilable" to maintain data integrity and avoid any bias in analysis.

```python
df["director"].fillna("No information avilable", inplace=True)
df["cast"].fillna("No information avilable", inplace=True)
```

* In the "country" column we fill in missing values with the mode (most frequently occurring value) to ensure consistency and minimize data loss.

```python
df["country"] = df["country"].fillna(df["country"].mode()[0])
```

* For the "rating" column, filling the missing values by the mode of the "type" of the show separately.

```python
# Finding the mode rating for movies and tv shows
movie_rating = df.loc[df["type"] == "Movie", "rating"].mode()[0]
tv_rating = df.loc[df["type"] == "TV Show", "rating"].mode()[0]
```

```python
# Filling missing rating values based on the type of content
df["rating"] = df.apply(lambda x: movie_rating if x["type"] == "Movie" and pd.isna(x["rating"])
              else tv_rating if x["type"] == "TV Show" and pd.isna(x["rating"])
              else x["rating"], axis=1 )
```

* For "duration" column, filling the missing values by the mode of the "type of show separately.

```python
# Finding the mode duration for movies and tv shows
movie_duration = df.loc[df["type"] == "Movie", "duration"].mode()[0]
tv_duration = df.loc[df["type"] == "TV Show", "duration"].mode()[0]
```

```python
# Filling missing rating values based on the type of content
df["duration"] = df.apply(lambda x: movie_duration if x["type"] == "Movie" and pd.isna(x["duration"])
              else tv_duration if x["type"] == "TV Show" and pd.isna(x["duration"])
              else x["duration"], axis=1 )
```

* After handling missing values in majority of column, the label "date_added" contain nan insignificant portion of the data, so to get rid of them drop the rows from the dataset.

```python
df.dropna(subset=["date_added"],inplace=True)
```

```python
df.isnull().sum()
```

* Finally we can see that there are no more missingvalue in the dataframe.

#**Data Handling**

* We convert the "date_added" column to datetime format using pd.to_datetime() for further analysis based on date related attributes.
*Next we extract additional attributes like month and year from "date_added" column to enhance our analysis capabilities.

```python
df["date_added"] = pd.to_datetime(df["date_added"])
```

#**Data Transformation**

* As we sow there are some column ("Director", "cast", "country" and "listed_in") having nested data, we transform them into separate dataframes. we split the values based on comma separator and created separate rows for each value.

```python
# spliting and expanding "director" column
df_director = df["director"].str.split(",",expand=True).stack()
df_director = df_director.reset_index(level=1, drop=True).to_frame("director")
df_director["show_id"] = df["show_id"]
df_director["type"] = df["type"]
```

```python
# spliting and expanding "cast" column
df_cast = df["cast"].str.split(",",expand=True).stack()
df_cast = df_cast.reset_index(level=1, drop=True).to_frame("cast")
df_cast["show_id"] = df["show_id"]
df_cast["type"] = df["type"]
```

```python
# spliting and expanding "country" column
df_country = df["country"].str.split(",",expand=True).stack()
df_country = df_country.reset_index(level=1, drop=True).to_frame("country")
df_country["show_id"] = df["show_id"]
df_country["type"] = df["type"]
```

```python
# spliting and expanding "listed_in" column
df_listed_in = df["listed_in"].str.split(",",expand=True).stack()
df_listed_in = df_listed_in.reset_index(level=1, drop=True).to_frame("listed_in")
df_listed_in["show_id"] = df["show_id"]
df_listed_in["type"] = df["type"]
```

* After compleating these data preparation steps, we have a clean and transforned dataset ready for further analysis.

**9) Genre Correlation Heatmap**
* Genres playa significant role in categorizing and organizing content on Netflix. Analyzing the correlation between genres can reveal interesting relationships between different types of content.  

* We create a genre dataframe to investigate genre correlation and fill it with zeros. By iterating over each row in the original dataframe, we update the genre dataframe based on the listed genres. we then create a corelation matrix using this genre data and vizualize it as a heatmap.

```python
# Extracting unique genre from the "listed_in" column
genres = df["listed_in"].str.split(",",expand=True).stack().unique()

# Create a new dataframe to store the genre data
genre_data = pd.DataFrame(index=genres, columns=genres, dtype=float)

# Fill the genre data DataFrame with zeros
genre_data.fillna(0, inplace=True)

# Iterate over each row in the original dataframe and update the genre dataframe
for _, row in df.iterrows() :
  listed_in = row["listed_in"].split(",")
  for genre1 in listed_in :
    for genre2 in listed_in :
      genre_data.at[genre1,genre2] += 1

# Creat a correlation matrix using the genre data
correlation_matrix = genre_data.corr()
```

```python
# Create the heatmap
plt.figure(figsize=(60,20))
sns.heatmap(correlation_matrix, annot=True, cmap="gist_stern_r")
# Customize the plot
plt.title("Genre Correlation Heatmap",fontsize=30)
plt.xticks(rotation=90)
plt.show()
```












* The heatmap demonstrates the correlation between different genres. By analyzing the heatmap, we can identify strong positive correlations between specific genres, such as TV Dramas and International TV Shows, Romantic TV Shows and International TV Shows.
